```python
import pandas as pd
from datetime import datetime
```

```python
annual = pd.read_csv("/home/aninotna/magister/tesis/justh2_pipeline/scripts/calliope_v5/data/extend_demand_h2_2100.csv", parse_dates=["year"]) 
annual["year"] = annual["year"].dt.year 
rows = []
```

```python
for _, r in annual.iterrows():
    y = int(r["year"])
    twh = float(r["demand_TWh"])
    idx = pd.date_range(f"{y}-01-01", f"{y}-12-31", freq="ME")
    mwh_m_total = -(twh * 1e6 / len(idx))  # TWh -> MWh y reparte por mes
    rows.append(pd.DataFrame({
        "timesteps": idx.strftime("%Y-%m-01"),
        "VALPO": mwh_m_total
    }))
```

```python
out = pd.concat(rows, ignore_index=True)
out.to_csv("../data/demand_h2_monthly_MWh.csv", index=False)
print("OK -> demand_h2_monthly_MWh.csv")
```
