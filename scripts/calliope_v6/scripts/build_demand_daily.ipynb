```python
import pandas as pd
from datetime import datetime
```

```python
annual = pd.read_csv("../data/demand_annual.csv", parse_dates=["year"]) 
annual["year"] = annual["year"].dt.year 
rows = []
```

```python
for _, r in annual.iterrows():
    y = int(r["year"])
    twh = float(r["demand_TWh"])
    idx = pd.date_range(f"{y}-01-01", f"{y}-12-31", freq="D")
    mwh_d_total = -(twh * 1e6 / len(idx))  # TWh -> MWh y reparte por dÃ­a
    rows.append(pd.DataFrame({
        "timesteps": idx.strftime("%Y-%m-%d"),
        "VALPO": mwh_d_total
    }))
```

```python
out = pd.concat(rows, ignore_index=True)
out.to_csv("../data/demand_h2_daily_MWh.csv", index=False)
print("OK -> demand_h2_daily_MWh.csv")
```
