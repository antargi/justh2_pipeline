```python
import pandas as pd
```

```python
year = 2025
idx = pd.date_range(f"{year}-01-01 00:00", f"{year}-12-31 23:00", freq="H")
```

```python
def cf_hour(h):
    if 6 <= h <= 18:
        return max(0, 0.9 - abs(h - 12) * 0.15)
    return 0.0
```

```python
years = range(2023, 2044)
rows = []
for y in years:
    idx = pd.date_range(f"{y}-01-01 00:00", f"{y}-12-31 23:00", freq="h")
    rows.append(pd.DataFrame({
        "timesteps": idx.strftime("%Y-%m-%d %H:%M"),
        "VALPO.pv": [round(cf_hour(ts.hour), 2) for ts in idx]
    }))
```

```python
out = pd.concat(rows, ignore_index=True)
out.to_csv("../data/pv_cf.csv", index=False)
print("OK -> pv_cf.csv (2023–2043)")
```
