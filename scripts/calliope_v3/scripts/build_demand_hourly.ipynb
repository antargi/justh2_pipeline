```python
import pandas as pd
from datetime import datetime
```

```python
annual = pd.read_csv("../data/demand_annual.csv", parse_dates=["year"]) 
annual["year"] = annual["year"].dt.year 
rows = []
```

```python
for _, r in annual.iterrows():
    y = int(r["year"])
    twh = float(r["demand_TWh"])
    idx = pd.date_range(f"{y}-01-01 00:00", f"{y}-12-31 23:00", freq="h")  # maneja bisiestos
    mwh_h_total = -(twh * 1e6 / len(idx))  # TWh -> MWh y reparte por hora
    rows.append(pd.DataFrame({
        "timesteps": idx.strftime("%Y-%m-%d %H:%M"),
        "VALPO": mwh_h_total
    }))
```

```python
out = pd.concat(rows, ignore_index=True)
out.to_csv("../data/demand_h2_MWh.csv", index=False)
print("OK -> demand_h2_MWh.csv")
```
