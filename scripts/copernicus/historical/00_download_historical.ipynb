```python
import cdsapi
import os

client = cdsapi.Client()

dataset = "projections-cmip6"

# Modelos y variables a iterar
models = ["access_cm2", "kace_1_0_g", "canesm5", "inm_cm4_8", "mpi_esm1_2_hr"]
variables = {
    "daily_minimum_near_surface_air_temperature": "tasmin"
}

# Plantilla base de la solicitud
request_template = {
    "temporal_resolution": "daily",
    "experiment": "historical",
    "month": [f"{m:02d}" for m in range(1, 13)],
    "day": [f"{d:02d}" for d in range(1, 32)],
    "year": [str(y) for y in range(1850, 2015)],
}

# Carpeta base
base_dir = "historical"
os.makedirs(base_dir, exist_ok=True)

for variable, short_name in variables.items():
    var_dir = os.path.join(base_dir, short_name)
    os.makedirs(var_dir, exist_ok=True)

    for model in models:
        try:
            request = request_template.copy()
            request["variable"] = variable
            request["model"] = model

            print(f"üîπ Descargando {short_name.upper()} del modelo {model}...")

            output_path = os.path.join(var_dir, f"{short_name}_{model}_historical_1850_2014.zip")

            client.retrieve(dataset, request).download(output_path)

            print(f"‚úÖ Guardado en: {output_path}\n")

        except Exception as e:
            print(f"‚ö†Ô∏è Error al descargar {short_name.upper()} para {model}: {e}\nContinuando con el siguiente...\n")
```
